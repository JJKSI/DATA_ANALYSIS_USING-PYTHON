## What is Normalization
### Normalization of attributes, also known as feature scaling or standardization, is the process of transforming numerical attributes in a dataset to a standard scale. This is done to ensure that all features contribute equally to the analysis and to avoid issues caused by differences in the scale of the attributes.

```python
# FOR NORMALIZATION PROCESS WE WILL REQUIRE THIS SKLEARN LIBRARY IN WHICH THERE ARE FUNCTIONS WHICH CAN HELP US TO NORMALIZE 
# THE VALUES WE WANT.

!pip install scikit-learn
```

#### For Normalization there are various methods provided from the sk-learn package. We can use Min-Max scaling, Z-score normalization. So in our case I am going with the Mix-MaxScaler method for normalizing our numerical Attribute.



#### So if I talk about Mix Max Scaler method It tries to scale the values of features between 0 and 1 it ensures that it represents all the values within that specific range(0,1). So if I talk about the statistics description after normalizaion then it will definetly different from the one before Normalization.
##### Min-Max scaling transforms the data by subtracting the minimum value and then dividing by the range (difference between the maximum and minimum values).

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

# Here we will be using the numerival attributes whose Missing values are already filled using median
numeric_attri_normalized = scaler.fit_transform(numeric_attri_filled)

# creating a data frame just for safer side 
numeric_attri_normalized_df = pd.DataFrame(numeric_attri_normalized,columns = numeric_attri_filled.columns)
```
