# feature understanding(univariate analysis)

```python
x=df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title='Top years introducced roller coaster')
x.set_xlabel('Year Introduced')
x.set_ylabel('Count')
```

```python
ax=df['speed_mph'].plot(kind='hist',title='Speed of roller coasters')
ax.set_xlabel('speed')
```

```python
df.plot(kind='scatter',
       x='speed_mph',
       y='height_ft',
       title='speed vs roller coaster height')
plt.show()# we can use to show outcome without extra data
```

```python
sns.scatterplot(
       x='speed_mph',
       y='height_ft',
       hue='Year_Introduced',           #hue can be use to compare with other variable for better visualization  
        data=df)
```

```python
#pair plot are nice to use
sns.pairplot(df,vars=['Year_Introduced','speed_mph','height_ft','Gforce_clean'],hue='Type_Main')#we can use pair plot when we want to compare multipe variables
plt.show()
```

```python
dfcor=df[['Year_Introduced','speed_mph','height_ft','Gforce_clean']].dropna().corr()
```
